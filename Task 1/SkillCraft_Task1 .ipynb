```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/ksoum/Downloads/1301- Dataful/population-projection-of-india-gender-and-age-group-wise-all-india-quinquennial-projected-population-2011-2036.csv')
```

```python
filtered_df = df[
    (df['year'] == 2021) &
    (df['state'] == 'All India') &
    (df['gender'] == 'Person') &
    (df['age_group'] != 'Total')
]
```

```python
total_population = filtered_df['value'].sum()
print("Total Population (2021):", total_population)
```

```python
age_groups = filtered_df['age_group']
population_values = filtered_df['value']
```

```python
plt.figure(figsize=(14, 7))
sns.set_style('whitegrid')

bar_plot = sns.barplot(x=age_groups, y=population_values,hue = age_groups, palette='viridis')
bar_plot.set_xticks(range(len(age_groups)))
bar_plot.set_xticklabels(age_groups, rotation=45)
plt.title("India Population Distribution by Age Group in 2021", fontsize=16)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Population (Thousands)", fontsize=14)

plt.text(0.5, 0.96,
         f"Total Population: {total_population:,.0f} thousands",
         ha='center', va='center',
         fontsize=14, color='darkred',
         transform=plt.gca().transAxes,
         weight='bold')

plt.tight_layout()
plt.show()
```

```python
filtered_df = df[
    (df["year"] == 2021) &
    (df["state"] == "All India") &
    (df["age_group"] != "Total")
]
```

```python
plt.figure(figsize=(16,8))
sns.set_style('whitegrid')
sns.barplot(
    data=filtered_df,
    x='age_group',
    y='value',
    hue='gender',
    palette="muted"
)
plt.xticks(rotation=45)
plt.title("India Population Distribution by Age Group and Gender (2021)", fontsize=16)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Population (Thousands)", fontsize=14)
plt.legend(title="Gender")

plt.text(0.5, 0.96,
         f"Total Population: {total_population:,.0f} thousands",
         ha='center', va='center',
         fontsize=14, color='darkred',
         transform=plt.gca().transAxes,
         weight='bold')

plt.tight_layout()
plt.show()
```
